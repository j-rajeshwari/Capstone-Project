```python
car_features_merged['max_torque_value']=car_features_merged['max_torque'].str.split('Nm@').str[0].astype(float)
car_features_merged['max_torque_rpm']=car_features_merged['max_torque'].str.split('Nm@').str[1].str.replace("rpm","").astype(float)
```

```python
car_features_merged['max_power_value']=car_features_merged['max_power'].str.split('bhp@').str[0].astype(float)
car_features_merged['max_power_rpm']=car_features_merged['max_power'].str.split('bhp@').str[1].str.replace("rpm","").astype(float)
```

```python
car_features_merged.head(1)
```

```python
car_features_merged['is_esc']=car_features_merged['is_esc'].str.lower().map({"yes":1,"no":0})
car_features_merged['is_adjustable_steering']=car_features_merged['is_adjustable_steering'].str.lower().map({"yes":1,"no":0})

car_features_merged['is_tpms']=car_features_merged['is_tpms'].str.lower().map({"yes":1,"no":0})
car_features_merged['is_parking_sensors']=car_features_merged['is_parking_sensors'].str.lower().map({"yes":1,"no":0})

car_features_merged['is_parking_camera']=car_features_merged['is_parking_camera'].str.lower().map({"yes":1,"no":0})
car_features_merged['is_rear_window_wiper']=car_features_merged['is_rear_window_wiper'].str.lower().map({"yes":1,"no":0})

car_features_merged['is_rear_window_washer']=car_features_merged['is_rear_window_washer'].str.lower().map({"yes":1,"no":0})
car_features_merged['is_rear_window_defogger']=car_features_merged['is_rear_window_defogger'].str.lower().map({"yes":1,"no":0})

car_features_merged['is_brake_assist']=car_features_merged['is_brake_assist'].str.lower().map({"yes":1,"no":0})
car_features_merged['is_power_door_locks']=car_features_merged['is_power_door_locks'].str.lower().map({"yes":1,"no":0})

car_features_merged['is_central_locking']=car_features_merged['is_central_locking'].str.lower().map({"yes":1,"no":0})
car_features_merged['is_power_steering']=car_features_merged['is_power_steering'].str.lower().map({"yes":1,"no":0})

car_features_merged['is_driver_seat_height_adjustable']=car_features_merged['is_driver_seat_height_adjustable'].str.lower().map({"yes":1,"no":0})
car_features_merged['is_day_night_rear_view_mirror']=car_features_merged['is_day_night_rear_view_mirror'].str.lower().map({"yes":1,"no":0})

car_features_merged['is_ecw']=car_features_merged['is_ecw'].str.lower().map({"yes":1,"no":0})
car_features_merged['is_speed_alert']=car_features_merged['is_speed_alert'].str.lower().map({"yes":1,"no":0})
```
